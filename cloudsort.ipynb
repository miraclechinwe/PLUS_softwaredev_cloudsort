Import a shapefile with the Area of Interest

```python
# Define the path to the shapefile containing the AOI
shapefile_path ='./Sample_data/Sample_AOI/AOI_Salzachauen_buffer_150m_WGS84_33N.shp'

# Load the shapefile using GeoPandas and convert it to Earth Engine object.
aoi = geemap.shp_to_ee(shapefile_path)

# Add AOI to the map.
Map.addLayer(aoi, {}, "AOI")
Map.centerObject(aoi, 10)
```

Create and filter Sentinel-2 image collection

```python
# Define the date range for filtering.
start_date = '2020-01-01'
end_date = '2021-01-01'

# Define the Sentinel-2 tile.
tile = '33UUP'

# Load the Sentinel-2 data and filter by date, bounds, and tile. 
# The collection is also sorted by images cloud cover and then each image is cliped by the aoi
dataset = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')\
    .filterDate(start_date, end_date)\
    .filterBounds(aoi)\
    .filter(ee.Filter.eq('MGRS_TILE', tile))\
    .sort('CLOUDY_PIXEL_PERCENTAGE')\
    .map(lambda image: image.clip(aoi))

dataset
```

```python
# Visualize the first 5 images of the S2 collection
num_images = 5
images_list = dataset.toList(num_images)
for i in range(num_images):
    image = ee.Image(images_list.get(i))
    # Define visualization parameters (adjust as needed)
    vis_params = {
        'bands': ['B4', 'B3', 'B2'],  # True color composite
        'min': 0,
        'max': 3000,
        'gamma': 1.4
    }
    # Add the image to the map
    Map.addLayer(image, vis_params, f'Image {i+1}')

Map
```
